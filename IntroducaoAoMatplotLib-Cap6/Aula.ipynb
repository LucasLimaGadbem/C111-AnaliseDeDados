# Introducao ao MatplotLib

importando as bibliotecas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt # o .pyplot Ã© para trabalhar com os grafico mais simples, pizza, barra, linha
```

Plot

```python
# o plot descreve as alteraÃ§Ãµes nos dados ao longo do tempo

# criando os eixos x e y
x = np.array([1,2,3,4])
y = x*2 #broadcasting #eixo y do grafico 1
y2 = x ** 2 #eixo y do grafico 2

plt.xlabel('Valores de x')
plt.ylabel('Valores de y')

#plotando  e formatando o grafico
#s - square(quadrado)       o - circulo
#: - pontilhada             - - linha
#r - red(vermelho)          b - azul
#linewidth - grossura da linha
#markersize - tamanho do quadrado
plt.plot(x, y, 's:r', x, y2, 'o-b',linewidth = 3, markersize = 20)
```

Subplots

```python
x = np.array([1,2,3,4])
y = x*2
y2 = x ** 2

plt.subplot(1,2,1) #uma linha, duas colunas, posicao 1
plt.title('FunÃ§Ã£o Linear')
plt.plot(x, y, 's:r')

plt.subplot(1, 2, 2) #uma linha, duas colunas, posicao 2
plt.title('FunÃ§Ã£o Exponencial')
plt.plot(x, y2, 'o-b')
```

GrÃ¡fico de barras

```python
# importando o dataset
df = pd.read_csv('paises.csv', delimiter = ';')

# extraindo as colunas
print(df.columns)

# extraindo os 5 maiores GDP do dataset
dfBiggestGDP = df.nlargest(5, 'GDP ($ per capita)')

eixo_x = dfBiggestGDP['Country']
eixo_y = dfBiggestGDP['GDP ($ per capita)']

# traÃ§ando o grÃ¡fico em barras
plt.bar(eixo_x, eixo_y, color = 'green')
```

GrÃ¡fico em torta

```python
#mostrando as colunas
#print(df.columns)

#mostrando os valores da Coastline
#print(df['Coastline (coast/area ratio)'])

#pegando os paises que nao tem costa
dfNoCoast = df[df['Coastline (coast/area ratio)'] == 0]

#analisando quantos nao tem costa
withCoast = len(df) - len(dfNoCoast)
withoutCoast = len(dfNoCoast)

# traÃ§ando o grÃ¡fico em torta
plt.pie(x = [withCoast, withoutCoast], 
        labels = ['% Paises com Costa', '% Paises sem Costa'],
        autopct = '%1.1f%%')
```
