## Exercicios

importando as bibliotecas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

importando os datasets

```python
paises = pd.read_csv('paises.csv', delimiter = ';')
space = pd.read_csv('space.csv', delimiter = ';')
```

mostrando as colunas dos datasets

```python
print(paises.columns)
print(space.columns)
```

1

```python
northAmerica = paises[paises['Region'].str.contains('NORTHERN AMERICA')]

nomes = northAmerica['Country']
mortalidade = northAmerica['Deathrate']
natalidade = northAmerica['Birthrate']

plt.title('Taxa de natalidade e mortalidade')
plt.xlabel('Paises')
plt.ylabel('Taxas')
plt.plot(nomes, natalidade, 'o-r')
plt.plot(nomes, mortalidade, 'o-b')
plt.legend(['Natalidade', 'Mortalidade'])
```

2

```python
empresaseua = space[space['Location'].str.contains('USA')]['Company Name'].unique()
empresaschina = space[space['Location'].str.contains('China')]['Company Name'].unique()

eixox = ['EUA', 'China']
eixoy = [len(empresaseua), len(empresaschina)]
plt.title('Empresas espaciais')
plt.xlabel('Paises')
plt.ylabel('Quantidade')
plt.bar(eixox, eixoy, color = 'green')
```

3

```python
roscosmos = space[space['Company Name'].str.contains('Roscosmos')]

sucesso = roscosmos[roscosmos['Status Mission'].str.contains('Success')]
fracasso = roscosmos[roscosmos['Status Mission'].str.contains('Failure')]

plt.pie(x = [len(sucesso), len(fracasso)], labels = ['Sucesso', 'Fracasso'], autopct='%1.1f%%')
```
